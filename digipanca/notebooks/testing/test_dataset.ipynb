```python
import os
import torch
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from src.data import PancreasDataset
```

```python
from src.data.dataset import RAW_DIR
```

```python
sample_dirs = [os.path.join(RAW_DIR, sd) for sd in os.listdir(RAW_DIR)]
print(sample_dirs)
```

```python
tendirs = sample_dirs[10:20]
print(tendirs)
```

```python
# dataset = PancreasDataset(tendirs)
```

```python
for dir in sample_dirs:
    print(dir)
    dataset = PancreasDataset([dir])
```

```python
# colores para segmentaci칩n
cmap = mcolors.ListedColormap(['black', 'green', 'purple', 'red', 'blue'])
boundaries = [0, 1, 2, 3, 4, 5]
norm = mcolors.BoundaryNorm(boundaries, cmap.N, clip=True)
```

```python
image, mask, patient_id = dataset[28]
# tensores a numpy
image_np = image.squeeze().cpu().numpy()
mask_np = mask.cpu().numpy()
```

```python
# Mostrar resultados
plt.figure(figsize=(12, 6))

# Imagen Original
plt.subplot(1, 2, 1)
plt.imshow(image_np, cmap="gray")
plt.title(f"Imagen Original - Paciente {patient_id}")

# Segmentaci칩n
plt.subplot(1, 2, 2)
plt.imshow(image_np, cmap="gray", alpha=0.0)  # Superponer la segmentaci칩n sobre la imagen
plt.imshow(mask_np, cmap=cmap, norm=norm, alpha=0.5)
plt.title(f"M치scara Segmentada - Paciente {patient_id}")

# Mostrar gr치fico
plt.show()
```
